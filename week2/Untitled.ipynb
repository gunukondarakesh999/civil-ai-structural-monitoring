```python
import pandas as pd

df = pd.read_excel("Concrete_Data.xls")  # or encoding="ISO-8859-1"
print(df.head())
df.describe()
```

```python
print(df.head())  # First 5 rows
```

```python
print("\nMissing values in each column:")
print(df.isnull().sum())
```

```python
# Step 5: Fill or drop missing values (if any)
df = df.dropna()  # You can also use df.fillna(value)
```

```python
# Step 6: Check duplicates
print("\nDuplicate rows count:", df.duplicated().sum())
```

```python
# Remove duplicates
df = df.drop_duplicates()
```

```python
print("\nData types of each column:")
print(df.dtypes)
```

```python
print(df.dtypes)
```

```python
# Step 8: Descriptive statistics
print("\nStatistics summary:")
print(df.describe())
```

```python
df.rename(columns={"Concrete compressive strength(MPa, megapascals) ": "Strength_MPa"}, inplace=True)
```
